```python
sns.pairplot(df)
```

**Rug Plot** The rugplot() is used to draw small bars along x-axis for each point in the dataset.

```python
sns.rugplot(df['Fare'])
```

**Bar Plot** The barplot() is used to display the mean value for each value in a categorical column, against a
numeric column.

```python
sns.barplot(x='Sex', y='Age', data=df)
```

**Bar Plot**  std aggregate function used from the numpy library to calculate the
standard deviation for the ages of male and female passengers.

```python
sns.barplot(x='Sex', y='Age', data=df, estimator=np.std)
```

**Count Plot**

```python
sns.countplot(x='Sex',data=df)
```

**Box Plot** The box plot is used to display the distribution of the categorical data in the form of quartiles.

```python
sns.boxplot(x='Sex', y='Age', data=df)
```

**Violin Plot** The violin plot allows us to display all the
components that actually correspond to the data point.

```python
sns.violinplot(x='Sex', y='Age', data=df)
```

**Strip Plot** The strip plot draws a scatter plot where one of the variables is categorical.

```python
sns.stripplot(x='Sex', y='Age', data=df)
```

**Swarm plot** The swarm plot is a combination of the strip and the violin plots. In the swarm plots, the points are
adjusted in such a way that they don't overlap.

```python
sns.swarmplot(x='Sex', y='Age', data=df)
```
