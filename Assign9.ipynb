**Box Plot**

```python
sns.boxplot(x='Sex', y='Age', data=df, hue="2urvived")
```

**Histogram** A histogram is a value distribution plot of numerical columns.

```python
plt.hist(df['Age'], bins=5)
```

**Dist Plot** Distplot is also known as the second Histogram because it is a slight improvement version of
the Histogram.

```python
sns.distplot(df['Age'])
```

**Pie Chart** The pie chart is also the same as the countplot, only gives us additional information about the
percentage

```python
df['Sex'].value_counts().plot(kind="pie", autopct="%.2f")
```
